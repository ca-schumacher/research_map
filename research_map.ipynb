Create the raw map

```python
f = folium.Figure(width=1000, height=500)
m = folium.Map(location=[30, 0], world_copy_jump=True, zoom_start=2, min_zoom = 1.7, tiles="openstreetmap").add_to(f)

marker_cluster = MarkerCluster(
    name='research_map',
    overlay=True,
    control=False
)

for i in range(0,len(html)):
    loca = data.iloc[i]['lon'], data.iloc[i]['lat']
    marker = folium.Marker(location=loca)
    iframe = branca.element.IFrame(html=html[i], width=500, height=200)
    popup = html[i]
    folium.Popup(popup).add_to(marker)
    marker_cluster.add_child(marker)

marker_cluster.add_to(m)

folium.LayerControl().add_to(m)
```
